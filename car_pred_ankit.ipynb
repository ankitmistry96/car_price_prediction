```python
# test train split 
from sklearn.model_selection import train_test_split
X_trian,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# using random forest regressor 
from sklearn.ensemble import RandomForestRegressor 
rf_random = RandomForestRegressor()
```

```python
# hyperparameters 
n_estimators=[int(x) for x in range(100,1200,100)]
max_features = ['auto','sqrt']
max_depth = [x for x in range(5,30,5)]
min_samples_split= [2,5,10,15,100]
min_samples_leaf = [1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# creating random grid
random_grid= {'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split':min_sample_split,
             'min_samples_leaf':min_sample_leaf}
print(random_grid)
```

```python
# use the random grid to select the best parameters
# first creating the base model to tune
rf= RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, 
                              param_distributions=random_grid,
                              scoring='neg_mean_squared_error',
                              n_iter=10,
                              cv=5,
                              verbose=2,
                              random_state=42,
                              n_jobs= 1)
```

```python
rf_random.fit(X,y)
```

```python
predictions = rf_random.predict(X_test)
```

```python
sb.distplot(y_test-predictions)
```
